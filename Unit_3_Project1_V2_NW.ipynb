```python
missing_values = ['s', 'na']
df = pd.read_csv ('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv', na_values = missing_values)
```

```python
df.head(10)
```

```python
drop_cols = ['Number Scoring CR', 'Percent Scoring CR']
df.drop (drop_cols, inplace = True, axis=1)
```

```python
df.head(30)
```

```python
df2= df[(df['School Name'] == 'Orchard Collegiate Academy')]
df2.head(30)
```

```python
agg_functions = {'Total Tested': 'sum', 'Number Scoring Below 65': 'sum', 'Number Scoring 65 or Above': 'sum', 'Number Scoring 80 or Above': 'sum',}
df3 = df2.groupby(df2['Year']).aggregate(agg_functions)
print(df3)
```

```python
df3.head()
```

```python
df3.groupby (['Year'])['Number Scoring Below 65']. sum(). plot(kind='bar', xlabel= 'Year', ylabel= 'Number of Students Scoring Below 65', title= 'Orchard Collegiate Academy_Below 65')
```

```python
df3.groupby (['Year'])['Number Scoring 65 or Above']. sum(). plot(kind='bar', xlabel= 'Year', ylabel= 'Number of Students Scoring 65 or Above', title= 'Orchard Collegiate Academy_65 or Above')
```

```python
df3.groupby (['Year'])['Number Scoring 80 or Above']. sum(). plot(kind='bar', xlabel= 'Year', ylabel= 'Number of Students Scoring 80 or Above', title= 'Orchard Collegiate Academy_80 or Above')
```

```python
df4= df[(df['School Name'] != 'Orchard Collegiate Academy') & (df['School Level'] == 'High school')]
df4.head(10)
```

```python
agg_functions = {'Total Tested': 'sum', 'Number Scoring Below 65': 'sum', 'Number Scoring 65 or Above': 'sum', 'Number Scoring 80 or Above': 'sum',}
df5 = df4.groupby(df4['Year']).aggregate(agg_functions)
print(df5)
```

```python
df5.head()
```

```python
df5.groupby (['Year'])['Number Scoring Below 65']. sum(). plot(kind='bar', xlabel= 'Year', ylabel= 'Number of Students Scoring Below 65', title= 'Other New York High Schools_Below 65', color='green')
```

```python
df5.groupby (['Year'])['Number Scoring 65 or Above']. sum(). plot(kind='bar', xlabel= 'Year', ylabel= 'Number of Students Scoring 65 or Above', title= 'Other New York High Schools_65 or Above', color='green')
```

```python
df5.groupby (['Year'])['Number Scoring 80 or Above']. sum(). plot(kind='bar', xlabel= 'Year', ylabel= 'Number of Students Scoring 80 or Above', title= 'Other New York High Schools_80 or Above', color='green')
```

```python
Orchard Collegiate Academy Regent Exam performance seems to show a unique trend.
While less students tested in 2017 compared to 2016, more students achieved scores of 65 or above and 80 or above year over year.

This trend is different compared to the trend of all other New York High Schools combined.
As less students tested year over year, less students achieved scores of 65 or above and 80 or above.

This suggests that Orchard Collegiate Academy might be an outlier among all other New York High Schools and prompts further investigations. 
```
