# Analysis Three - Determining When People are Enrolling into Competitor Plans  to Understand When Should the Campaign Launch?

```python
# Competitor enrollment trends before the COVID-19 pandemic (2017, 2018, 2019)
```

```python
df17= df2[(df2["Economic Region"] =="Long Island") & (df2["Eligibility Year"] == 2017) & (df2["Plan Name"] != "HEALTHFIRST")]
agg_functions = {"Number of Enrollees": "sum"}
df17.groupby(df17["Eligibility Month"]).aggregate(agg_functions)
```

```python
df18= df2[(df2["Economic Region"] == "Long Island") & (df2["Eligibility Year"] == 2018) & (df2["Plan Name"] != "HEALTHFIRST")]
agg_functions = {"Number of Enrollees": "sum"}
df18.groupby(df18["Eligibility Month"]).aggregate(agg_functions)
```

```python
df19= df2[(df2["Economic Region"] == "Long Island") & (df2["Eligibility Year"] == 2019) & (df2["Plan Name"] != "HEALTHFIRST")]
agg_functions = {"Number of Enrollees": "sum"}
df19.groupby(df19["Eligibility Month"]).aggregate(agg_functions)
```

```python
df17.groupby (["Eligibility Month"])["Number of Enrollees"]. sum(). plot(color="yellow")
df18.groupby (["Eligibility Month"])["Number of Enrollees"]. sum(). plot(color="red")
df19.groupby (["Eligibility Month"])["Number of Enrollees"]. sum(). plot(kind="line", xlabel="Month", ylabel="Number Enrollees", title="Number of Competitor Enrollees by Month in Long Island (Pre-Pandemic)", color="blue")
plt.legend(["2017","2018","2019"], loc="upper left")
```

```python
df20.groupby (["Eligibility Month"])["Number of Enrollees"]. sum(). plot(color="yellow")
df21.groupby (["Eligibility Month"])["Number of Enrollees"]. sum(). plot(color="red")
df22.groupby (["Eligibility Month"])["Number of Enrollees"]. sum(). plot(kind="line", xlabel= "Month", ylabel= "Number Enrollees", title= "Number of Competitor Enrollees by Month in Long Island (During COVID-19 Pandemic)", color="blue")
plt.legend(["2020","2021","2022"], loc="upper right")
```

# Analysis Three Observations:

# Competitors see most parents and legal guardians in Long Island enroll their children toward the second half of the year during pre-pandemic years (2017-2019). However, during years where the COVID-19 pandemic occurred (2020-2022) competitors saw the most enrollment toward the first half of the year.

# As we emerge out of the pandemic (2023 and 2024) we are seeing CHP enrollment trends shift. Data suggests that competitors are seeing greater enrollment toward the second half of the year once again.

# Conclusion

# Since the majority of market enrollment and competitor enrollment are from children between the ages of "13-19", Healthfrist should prioritize targeting parents of teenagers (13-19). This potentially lessens enrollment to competitor plans and stimulates its own growth, increasing market share. Parents of children/pre-teens (6-12) can be a viable secondary audience.

# Enrollment trends suggest that Healthfirst should run its campaign in Q1 and/or Q2. This helps influence the market before potential peaks in enrollment among competitors.
