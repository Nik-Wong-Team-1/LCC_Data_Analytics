# Importing Tools and Data

```python
import seaborn as sns
%matplotlib inline 
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/niteen11/DataAnalyticsAcademy/master/Python/dataset_diabetes/diabetic_data.csv')
```

# Data Cleaning

```python
drop_cols = ['weight', 'readmitted']
df.drop (drop_cols, inplace = True, axis=1)
```

```python
df.head()
```

```python
df1 = df[df['gender'] != 'Unknown/Invalid']
```

```python
df1.groupby (by= 'gender'). count()
```

# Filtering Desired Data based on Gender and Reported Diabetes Medication Use

```python
df2= df1[['gender','diabetesMed']]
df2.head()
```

```python
df3= df2[(df2['gender'] == 'Female')]
df3
```

```python
df3.groupby (by= 'diabetesMed'). count()
```

```python
df4= df2[(df2['gender'] == 'Male')]
df4
```

```python
df4.groupby (by= 'diabetesMed'). count()
```

# Data Visualization and Compairing Data Trends

```python
df3.groupby (['diabetesMed']). count(). plot(kind='bar', xlabel='Responses', ylabel='Count', title= 'Females and Diabetes Medication Use', color='pink', legend=False)
df4.groupby (['diabetesMed']). count(). plot(kind='bar', xlabel='Responses', ylabel='Count', title= 'Males and Diabetes Medication Use', color='blue', legend=False)
```

# Key Observations
Diabetes medication use trends suggest that diabetes management is highly relevant for both genders. From a paid social media perspective, this indicates that we should target both men and women with Diabetes Awarness Month content.

```python
df9.groupby (by= 'age'). count()
```

# Data Visualization and Compairing Data Trends

```python
df7.groupby (['age']). count(). plot(kind='bar', xlabel='Age', ylabel='Count', title= 'Females on Diabetes Medication (By Age)', color='pink', legend=False)
df9.groupby (['age']). count(). plot(kind='bar', xlabel='Age', ylabel='Count', title= 'Males on Diabetes Medication (By Age)', color='blue', legend=False)
```

# Additional Key Observations
Further analysis shows that female and male audiences who are between 70-80 years of age report the highest use of diabetes medication. This suggests that diabetes management is the most relevant for audiences between 70-80 years old. From a marketing perspective, this indicates that we should prioritize sharing Diabetes Awarness Month content with men and women who are 70-80 years old either through their caregivers or their senior living facilities.
