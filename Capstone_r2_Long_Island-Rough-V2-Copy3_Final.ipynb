# Analysis Three - Identifying Who is Enrolling into Competitors Plans

```python
# All enrollees enrolling into a competitor plan in Long Island (2017-2024)
# By race
```

```python
df_AD= df2[(df2["Economic Region"] == "Long Island") & (df2["Eligibility Year"] > 2016) & (df2["Plan Name"] != "HEALTHFIRST")]
agg_functions = {"Number of Enrollees": "sum"}
df_AD.groupby(df_AD["Race"]).aggregate(agg_functions)
```

```python
df_AD.groupby (["Race"])["Number of Enrollees"]. sum(). plot(kind="bar", xlabel= "Race", ylabel= "Number Enrollees", title= "Enrollees by Race", color="green")
```

# Analysis Three Observations

#  Based data from 2017 to 2024, Long Island competitors see their highest Child Health Plus enrollment from parents or legal guardians who identify their child as “other."

# Enrollment from parents or legal guardians who identify their child as "white" follow behind.

# Conclusion

# As Healthfirst leads in enrollments in 2024 thus far, it is suggested that our campaign's objective should aim to maintain Healthfirst’s position in Long Island as the health insurance plan with the highest enrollment.
# Enrollment trends suggest that we should launch a campaign during the first half of the year before competitors are likely to see the height of their enrollment toward the latter half of the year. This would ideally be a campaign that draws audience attention away from our competitors while encouraging enrollment to our plan.

# As competitors see their highest enrollment from parents or legal guardians who identify their child as either “other" or "white", our campaign should prioritize targeting audiences of the "other" demographic and leverage white audiences as a secondary target. This would potentially lessen enrollment to competitor plans while stimulating enrollment for Healthfirst.
